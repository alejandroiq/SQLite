```python
import sqlite3
import pandas as pd
conn = sqlite3.connect('data (3).sqlite')
```

```python
# One to One

q = """
SELECT firstName, lastName, email
FROM employees
WHERE jobTitle = 'Sales Rep'
;
"""
df = pd.read_sql(q, conn)
df
```

```python
# Sales Rep Employees with the city where they are located

q = """
SELECT firstName, lastName, email, city
      FROM employees
      JOIN offices
      USING(officeCode)
      WHERE jobtitle = 'Sales Rep'      
      GROUP BY city
"""
pd.read_sql(q, conn)
```

```python
q = """
SELECT productLine, productVendor, textDescription, productDescription
      FROM productlines
      JOIN products
      USING(productLine)
      
"""
pd.read_sql(q, conn)
```
