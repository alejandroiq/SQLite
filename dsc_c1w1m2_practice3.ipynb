# Practice for Grouping Data with SQL

Import the germane libraries and connect to the database.

```python
# Run this code

import pandas as pd
import sqlite3

conn = sqlite3.connect('data (3).sqlite')
pd.read_sql("""
SELECT *
  FROM products;
""", conn)
```

1. Select the product line
*   Count the number of each product line
*   Give the count the alias "count"
*   Sort descending by count

```python
# Group by: This clause is a MUST with aggreated functions

pd.read_sql("""
SELECT productLine,
        COUNT(*) AS my_count
  FROM products
  GROUP BY productLine
  ORDER BY my_count DESC;
""", conn)
```

2. Select the product line
- Determine the average by Price by product line
- Give the average the alias "avgPrice"
- Sort descending by avgPrice

```python
# Group by

pd.read_sql("""
SELECT productLine, AVG(CAST(buyPrice AS INTEGER)) AS avg_buyPrice
  FROM products
  GROUP BY productLine
  ORDER BY avg_buyPrice DESC;
""", conn)
```

3. Select the product line
- Find the minimum of the MSRP as minMSRP
- Find the maximum of the MSRP as maxMSRP

```python
# Group by

pd.read_sql("""
SELECT productLine, MIN(buyPrice) AS min_buyPrice, MAX(buyPrice) AS max_buyPrice
  FROM products
  GROUP BY productLine
""", conn)
```

4. Repeat (3), but only select the rows where the MSRP is $50 or more.

```python
# Group by & where

pd.read_sql("""
SELECT productLine, MIN(MSRP) AS min_MSRP, MAX(MSRP) AS max_MSRP
  FROM products
  WHERE MSRP >= 50
  GROUP BY productLine
""", conn)
```

5. Repeat (2), but only select the product lines with an average price greater than $50.  

```python
# Group by & having

pd.read_sql("""
SELECT productLine, AVG(buyPrice) as avg_buyPrice
  FROM products
  GROUP BY productLine
  HAVING avg_buyPrice >= 50
  ORDER BY avg_buyPrice DESC
""", conn)
```

6. Select the product line
- Find the average buy price as 'buyPrice' and the average MSRP as 'avgMSRP'
- Only select MSRP greater than equal to $50
- Only select avgPrice greater than equal to $50
- Group by by product line and order by the ascending average price

```python
# Group by & where & having

pd.read_sql("""
SELECT productLine, AVG(buyPrice) as avg_buyPrice, AVG(MSRP) as avg_MSRP
  FROM products
  WHERE MSRP >= 50
  GROUP BY productLine
  HAVING avg_MSRP >= 50
  ORDER BY avg_MSRP ASC
""", conn)
```

```python
pd.read_sql("""
SELECT productLine, AVG(buyPrice) AS avgPrice, AVG(MSRP) AS avgMSRP
 FROM products
 WHERE MSRP >= 50
 GROUP BY productLine
 HAVING avgPrice >= 50
 ORDER By avgPrice ASC
""", conn)
```
